ヘルプについて。関数のヘルプは`?mean`とか。関数名がわからないときは関連する語句を`??`を冒頭につけて検索する。  
予約されている語句や記号は引用符で囲む。  
例:  
```
?mean                #opens the help page for the mean function
?"+"                 #opens the help page for addition
?"if"                #opens the help page for if, used for branching code
??plotting           #searches for topics containing words like "plotting"
??"regression model" #searches for topics containing phrases like this
```
\#はコメントです。

関数の`help()`と`help.search()`はそれぞれ`?`と`??`と同じことです。  
`help()`と`help.search()`は引数を引用符で囲む必要があります。
```
help("mean")
help("+")
help("if")
help.search("plotting")
help.search("regression model")
```
関数`apropos()`は引数に一致する変数や関数を探します。  
これは自分が作った変数を探すのにも役立ちます。

例えば、次のように`a_vector`という変数を作ったとしましょう。

````r
a_vector <- c(1, 3, 6, 10)
````

`apropos("vector")`とすると、`vector`という文字列を含む変数や関数が返されます。

````r
apropos("vector")
````

`vector`の場合、`a_vector'以外はRにもともとある関数です。  

ところで、`apropos()`関数は引数に正規表現が使えます。  
正規表現ってなに？ というひとはさておいて、正規表現を使い慣れている人は便利に使えるかもしれません。

````r
apropos("z$")       # zで終わる変数を検索
apropos("[4-9]")    # 4から9までの数字を含む変数を検索
````
